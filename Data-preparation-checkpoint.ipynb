```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import spectrogram
import pywt
import os
from tqdm import tqdm
```

```python
fs = 5000
duration = 1
samples_per_class = 380
t = np.linspace(0, duration, fs)
base_dir = "dataset"
```

```python
classes = ["Non_Islanding", "Islanding"]

for cls in classes:
    os.makedirs(f"{base_dir}/{cls}", exist_ok=True)
```

```python
# THD CALCULATION

def calculate_thd(signal):
    fft_vals = np.abs(np.fft.fft(signal))
    freqs = np.fft.fftfreq(len(signal), 1/fs)

    fundamental_idx = np.argmin(np.abs(freqs - 50))
    V1 = fft_vals[fundamental_idx]

    harmonic_power = np.sum(fft_vals[2*fundamental_idx:10*fundamental_idx]**2)
    thd = np.sqrt(harmonic_power) / (V1 + 1e-10)
    return thd
```

```python
# NEGATIVE SEQUENCE DERIVATIVE

def compute_dvneg_dt(Va, Vb, Vc):
    a = np.exp(1j * 2*np.pi/3)
    Vneg = (Va + a**2 * Vb + a * Vc) / 3
    Vneg_mag = np.abs(Vneg)
    dvneg_dt = np.gradient(Vneg_mag)
    return dvneg_dt
```

```python
# SIGNAL GENERATORS

def generate_non_islanding():
    Va = np.sin(2*np.pi*50*t)
    Vb = np.sin(2*np.pi*50*t - 2*np.pi/3)
    Vc = np.sin(2*np.pi*50*t + 2*np.pi/3)

    Va += 0.05*np.sin(2*np.pi*150*t)
    Vb += 0.05*np.sin(2*np.pi*150*t)
    Vc += 0.05*np.sin(2*np.pi*150*t)

    Ia = Va + 0.02*np.random.randn(len(t))
    Ib = Vb + 0.02*np.random.randn(len(t))
    Ic = Vc + 0.02*np.random.randn(len(t))

    return Va, Vb, Vc, Ia, Ib, Ic
```

```python
def generate_islanding():
    Va = np.sin(2*np.pi*50*t)
    Vb = np.sin(2*np.pi*50*t - 2*np.pi/3)
    Vc = np.sin(2*np.pi*50*t + 2*np.pi/3)

    # imbalance
    Va[int(0.5*fs):] *= 0.6
    Vb[int(0.5*fs):] *= 0.5
    Vc[int(0.5*fs):] *= 0.8

    # frequency shift
    Va[int(0.6*fs):] = np.sin(2*np.pi*52*t[int(0.6*fs):])

    # harmonics
    Va += 0.2*np.sin(2*np.pi*150*t)
    Vb += 0.15*np.sin(2*np.pi*150*t)
    Vc += 0.1*np.sin(2*np.pi*150*t)

    Ia = Va + 0.1*np.random.randn(len(t))
    Ib = Vb + 0.1*np.random.randn(len(t))
    Ic = Vc + 0.1*np.random.randn(len(t))

    return Va, Vb, Vc, Ia, Ib, Ic
```

```python
import matplotlib
matplotlib.use('Agg')   # IMPORTANT: Non-interactive backend

import gc

def save_images(signal, label, idx):

    # Normalize
    signal = (signal - np.mean(signal)) / (np.std(signal) + 1e-10)

    # Convert to float32 (memory saver)
    signal = signal.astype(np.float32)

    # ------------------- Spectrogram -------------------
    f, t_spec, Sxx = spectrogram(signal, fs)

    fig = plt.figure(figsize=(3,3))
    plt.pcolormesh(t_spec, f, 10*np.log10(Sxx + 1e-10))
    plt.axis('off')
    plt.savefig(f"{base_dir}/{label}/sample_{idx}_spectrogram.png",
                bbox_inches='tight',
                pad_inches=0,
                dpi=80)
    plt.close(fig)

    # ------------------- Scalogram -------------------
    scales = np.arange(1, 64)   # Reduced from 128 → HUGE memory saver

    coeffs, _ = pywt.cwt(signal, scales, 'morl', sampling_period=1/fs)

    fig = plt.figure(figsize=(3,3))
    plt.imshow(np.abs(coeffs).astype(np.float32),
               aspect='auto',
               cmap='jet')
    plt.axis('off')
    plt.savefig(f"{base_dir}/{label}/sample_{idx}_scalogram.png",
                bbox_inches='tight',
                pad_inches=0,
                dpi=80)
    plt.close(fig)

    del coeffs
    gc.collect()   # FORCE memory cleanup
```

```python
for cls in classes:
    print(f"Generating {cls} samples...")
    for i in tqdm(range(samples_per_class)):

        if cls == "Non_Islanding":
            Va, Vb, Vc, Ia, Ib, Ic = generate_non_islanding()
        else:
            Va, Vb, Vc, Ia, Ib, Ic = generate_islanding()

        # Features
        dvneg_dt = compute_dvneg_dt(Va, Vb, Vc)
        voltage_thd = calculate_thd(Va)
        current_thd = calculate_thd(Ia)

        # Combine features into dynamic signal
        feature_signal = dvneg_dt + voltage_thd + current_thd

        save_images(feature_signal, cls, i)

print("Dataset generation complete.")
```
